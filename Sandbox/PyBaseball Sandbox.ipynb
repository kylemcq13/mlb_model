```python
# retrieve each player's table

cole = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 543037)
lynn = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 458681)
karinchak = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 675916)
bauer = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 545333)
kershaw = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 477132)
rodon = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 607074)
ohtani = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 660271)
darvish = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 506433)
bieber = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 669456)
degrom = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 594798)
scherzer = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 453286)
woodruff = statcast_pitcher('2021-01-01', '2021-07-31', player_id = 605540)
```

```python
df = pd.concat([cole, lynn, karinchak, bauer,
               kershaw, rodon, ohtani, darvish,
               bieber, degrom, scherzer, woodruff])
```

```python
spin = df[['player_name', 'game_date', 'pitch_type', 'release_spin_rate', 'release_extension']]
spin.head()
```

```python
spin['player_name'].unique()
```

```python
spin['game_date'] = pd.to_datetime(spin['game_date'])
```

```python
spin = (spin.groupby(['player_name', 'game_date', 'pitch_type'])
                   ['release_spin_rate', 'release_extension']
                   .mean()
                   .sort_values(by='game_date')
                   .reset_index())
```

```python
karinchak = spin.loc[spin['player_name']=='Karinchak, James']
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

def plot_spin(player):
    plt.figure(figsize=(20,20))
    pitcher = spin.loc[spin['player_name']==player]
    sns.lineplot(data=pitcher, x='game_date', y='release_spin_rate', hue='pitch_type')
    
plot_spin('Scherzer, Max')
```
